```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
import time
import numpy as np
import pandas as pd
```

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

site= "https://ingatlan.com/lista/elado+lakas?page=1"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page)
```

```python
sitenum = soup.find_all("div", class_="pagination__page-number")
sitenum = sitenum[0]
sitenum = ''.join(i for i in sitenum if i not in '"')
sitenum = sitenum.replace(' 1 / ','')
sitenum = sitenum.replace('oldal','')
sitenum = sitenum.replace(' ','')
sitenum = int(sitenum)
print(sitenum)
```

```python
i = 1
links = []
while i <= sitenum:
    site= "https://ingatlan.com/lista/elado+lakas?page=" + str(i)
    req = Request(site,headers=hdr)
    page = urlopen(req)
    soup = BeautifulSoup(page)

    for link in soup.findAll('a'):
        links.append(link.get('href'))

    links = [i for i in links if i]
    substring = ["/elado+lakas/"]
    matching = [s for s in links if any(xs in s for xs in substring)]
    time.sleep(2.5)
    print("kész oldalak: " + str(i))
    i = i + 1
```

```python
matching = np.array(matching)
```

```python
df = pd.DataFrame()
df['link']=pd.Series(matching)
```

```python
df.to_csv("links.csv")
```
